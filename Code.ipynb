```python
import pandas as pd
from sklearn.model_selection import train_test_split


canada_sars = pd.read_csv('sars_canada.csv')
canada_covid = pd.read_csv('covid_canada.csv')

display(canada_sars)
display(canada_covid)
```

```python
# Splits the two data frames into training and test data, x and y values are not defined yet

sars_train, sars_test = train_test_split(canada_sars, test_size=.2, random_state = 8)
covid_train, covid_test = train_test_split(canada_covid, test_size=.2, random_state=8)

display(sars_train)
display(sars_test)
display(covid_train)
display(covid_test)
```
